#Executive Summary#

##What is your goal?##

My goal is to accurately predict the strength of a persons hand based on the line of betting they took throughout the hand. My target variable is hand strength, a number between 0 - 1, 1 being the best possible hand with the board, and 0 being the worst possible hand. This number is a percentile, and is calculated by iterating through all possible 2 card hands (removing cards on the board and in hand) and comparing the two hands. if it beats 100 percent of all other hands, it gets a hand strength of 1, if it beats 0, it gets 0. 
    For example, if the board is [As, 2d, 3h], the best possible hand you could make with this board are all the combinations of [4x, 5x] and the worst hand you could have would be all combinations of [4x, 6x]. Because of the nature of poker and how players play the game, and how pots get exponentially bigger with stronger hands, squaring the hand strengths seemed like a good idea to get closer to a linear relationship with amount of money put in the pot. This can be clearly seen through the scatter plot of bets and hand strength.

##What are your metrics?##

My metrics are not that good, my model is scoring within the .43 and .45 range, with a median difference between predicted and actual of about .12 in the hand strength ^2 metric. This could be the difference between a pair and three of a kind, and we need better accuracy to correctly predict for our bot. There are many different types of poker players, and generalizing over all of them is not very useful. I am working on a clustering model using non-showdown stats to group players into more useful groupings and hopefully create a bunch of models that can be used for each player type to get better accuracy.

##What were your findings?##

I am finding that generalizing all poker players into one 'average' player to predict might be too inaccurate to get reliable predictions. I am working on building a clustering model to cluster players into groups, and then have individual models for each of those groups. When the poker robot encounters a new player, it will have a 'starter stats' with a generalized (bad) predictions, but as it plays against a player, track their stats, update the stats, then put them into their appropriate group and change regression models accordingly.

##What risks/limitations/assumptions affect these findings?##

Because I can only see hands that make it to showdown, I have a biased sample. I probably cannot use anything related to frequency of these hand strengths to predict other things, because most hands do not make it to showdown. There is still useful data here though, and I can predict the hand strength being REPRESENTED, and then try and classify bluffs better and see if I can accurately predict bluffs with a logistic model. It would actually be better to get a percentage too because I can use those percentages to balance actions and keep the bot unpredictable. For example, if the bot raises every time it thinks an opponent is bluffing, the opponent can just adapt by playing their big hands like they are bluffing. If we get a PERCENTAGE probability that someone is bluffing, we can tell the bluff to randomly choose based on that probability to raise or to fold, but making sure to keep a positive expected value based on that percentage.
